```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('C://Users//USER//Documents//users.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
# Show the ages with the highest income
age_income = df[['Age', 'Total Income']].sort_values(by='Total Income', ascending=False).head(10)


age_income
```

```python
# Show the gender distribution across spending categories
gender_spending = df.groupby('Gender').mean()
gender_spending
```

```python
# Ages with the highest income
plt.figure(figsize=(10, 6))
sns.barplot(x='Age', y='Total Income', data=age_income, palette='viridis')
plt.title('Top 10 Ages with the Highest Income')
plt.xlabel('Age')
plt.ylabel('Total Income')
plt.show()
```

```python
#Gender distribution across spending categories
plt.figure(figsize=(12, 8))
gender_spending.plot(kind='bar')
plt.title('Gender Distribution Across Spending Categories')
plt.xlabel('Gender')
plt.ylabel('Average Spending')
plt.show()
```

```python
#  Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=10, kde=True, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#  Total Income distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Total Income'], bins=10, kde=True, color='green')
plt.title('Total Income Distribution')
plt.xlabel('Total Income')
plt.ylabel('Frequency')
plt.show()
```

```python
# Utilities spending by age
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Utilities', hue='Gender', data=df, palette='deep')
plt.title('Utilities Spending by Age')
plt.xlabel('Age')
plt.ylabel('Utilities')
plt.show()
```

```python
#  Entertainment spending by age
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Entertainment', hue='Gender', data=df, palette='muted')
plt.title('Entertainment Spending by Age')
plt.xlabel('Age')
plt.ylabel('Entertainment')
plt.show()
```

```python
#Correlation heatmap of spending categories
plt.figure(figsize=(10, 8))
corr = df[['Total Income', 'Utilities', 'Entertainment', 'School Fees', 'Shopping', 'Healthcare']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap of Spending Categories')
plt.show()
```
